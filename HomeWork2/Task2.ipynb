```python
abMore10 = data[data['absences']>10]
abLess10 = data[data['absences']<=10]
if abMore10['G3'].mean()<abLess10['G3'].mean():
    print('true')
else: print('false')
```

```python
dalcMore3 = data[data['Dalc']>3]
dalcLess3 = data[data['Dalc']<=3]
if dalcMore3['G3'].mean()<dalcLess3['G3'].mean():
    print('Dalc affects on G3')
else: print('Dalc does not affect on G3')
    
yes = data[data['higher']=='yes']
no = data[data['higher']=='no']
if yes['G3'].mean()<no['G3'].mean():
    print('Desire to learn affects on G3')
else: print('Desire to learn does not affect on G3')
    
yes = data[data['famsup']=='yes']
no = data[data['famsup']=='no']
if yes['G3'].mean()<no['G3'].mean():
    print('Support from family affects on G3')
else: print('Support from family does not affect on G3')
```
